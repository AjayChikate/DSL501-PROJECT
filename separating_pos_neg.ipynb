```python
import pandas as pd

DATA_PATH = "/content/pairs_for_pmponly.csv"   # CHANGE THIS

df = pd.read_csv(DATA_PATH)

print("Total rows:", len(df))
print(df.head())
```

```python
# Separate
df_pos = df[df["is_positive"] == 1].reset_index(drop=True)
df_neg = df[df["is_positive"] == 0].reset_index(drop=True)

print("Positive samples:", len(df_pos))
print("Negative samples:", len(df_neg))

# Save temporarily (optional)
df_pos.to_csv("positive_only.csv", index=False)
df_neg.to_csv("negative_only.csv", index=False)
```
